<h2>1. Jelaskan apa itu covariance dan cara menghitungnya!</h2>

Covariance adalah ukuran yang digunakan untuk melihat kecenderungan hubungan antara dua variable atau lebih untuk bervariasi bersama. Hasilnya dapat berupa simpangan positif atau negatif.

Cara menghitungnya dengan persamaan yang di dalamnya mengandung variabel dependence dan independence. Variabel dependence ini bergantung pada variabel-variabel independence yang ada.

\begin{equation*}
Cov(X , Y ) = \frac {1}{n} \sum dx_i dy_i
\end{equation*}

*source : Ref_2_Allen B. Downey - Think Stats-O'Reilly Media (2014)

<h2>3. Buat kode Python yang membaca file CSV yang diberikan, buat scatterplot dan hitung correlation Mileage vs Price! (boleh pakai Numpy)</h2>

```python
# import library
import pandas as pd
import seaborn as sns
import numpy as np

pd.set_option('display.max_rows', None)

# initial and read dataset
car_sales_data = pd.read_csv('Dataset Penjualan Mobil.csv', names=['Brand', 'Price', 'Body', 'Mileage', 'EngineV', 'Engine Type', 'Registration', 'Year', 'Model'])

# print dataset in table
car_sales_data
```

```python
# print dataset with scatterplot
sns.scatterplot(x='Mileage', y='Price', data=car_sales_data, hue='Brand')
```

Berdasarkan grafik di atas dapat dilihat kemungkinan membentuk hubungan non-linier

```python
# Calculate Correlation

# select and init dataframe mileage and price
df = car_sales_data[['Mileage', 'Price']]

# print dataframe
df
```

```python
# calculate correlation with method spearman
# use method spearman, because the shape of the probable graph is non-linear
df.corr('spearman')
```

<h2>4. Buat kode Python yang menghitung covariance Mileage vs Price! (hanya boleh pakai Pandas)</h2>

```python
# Calculate and print covariance
df.cov()
```
